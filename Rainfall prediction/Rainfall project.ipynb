```python
data.drop(['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'],axis=1,inplace=True)
data
```

```python
def remove_outliers(column,data,max_p:float,min_p:float)->None:
    maxima=data[column].quantile(max_p)
    minima=data[column].quantile(min_p)
    data[column][data[column]>maxima]=maxima
    data[column][data[column]<minima]=minima
    
```

```python
remove_outliers(column='Jan-Feb',data=data,max_p=0.85,min_p=0.1)
remove_outliers(column='Mar-May',data=data,max_p=0.85,min_p=0.1)
remove_outliers(column='Jun-Sep',data=data,max_p=0.85,min_p=0.1)
remove_outliers(column='Oct-Dec',data=data,max_p=0.85,min_p=0.1)
```

```python
scaler=MinMaxScaler()
```

```python
data['Jan-Feb']=scaler.fit_transform(data['Jan-Feb'].values.reshape(-1,1))
data['Mar-May']=scaler.fit_transform(data['Mar-May'].values.reshape(-1,1))
data['Jun-Sep']=scaler.fit_transform(data['Jun-Sep'].values.reshape(-1,1))
data['Oct-Dec']=scaler.fit_transform(data['Oct-Dec'].values.reshape(-1,1))
data['ANNUAL']=scaler.fit_transform(data['ANNUAL'].values.reshape(-1,1))
```

```python
encoder=LabelEncoder()
```

```python
encoded_values=encoder.fit_transform(data['SUBDIVISION'])
data['SUBDIVISION']=encoded_values

decoded_values=encoder.inverse_transform(encoded_values)
```

```python
mappings={}
for i in range (data.shape[0]):
    mappings[decoded_values[i]]=encoded_values[i]
```

```python
def generate_probs(annual):
    p75,p50=data['ANNUAL'].quantile(0.75),data['ANNUAL'].quantile(0.5)
    if np.greater(annual,p75):
        return 'Strong'
    elif np.greater_equal(annual,p50)and np.less(annual,p75):
        return 'Medium'
    else:
        return 'Low'
```

```python
data['Probability']=data.apply(lambda x: generate_probs(x['ANNUAL']),axis=1)
```

```python
x=data.drop(['Probability','ANNUAL','YEAR'],axis=1)
y=data['Probability']
```

```python
x
```

```python
model=SVC(kernel='linear',gamma='auto')
```

```python
model.fit(x,y)
```

```python
preds=model.predict(x)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,preds)
```

```python
data['PREDS']=preds
```

```python
fig,axis =plt.subplots(1,2)
plt.subplots_adjust(right=1.5)

axis[0].plot(data['Probability'],'blue')
axis[0].set(title='Actual Values')

axis[1].plot(data['PREDS'],'green')
axis[1].set(title='Predicted Values')
plt.show()
```
